```python
import numpy as np
# numpy 모듈을 가져오고 np라는 이름으로 사용을 한다.
# numpy는 선형대수연산, 푸리에 변환 등 수학함수를 포함
```

```python
x=np.array([[1,2,3],[4,5,6]])
print("x:\n",x)
# np 모듈의 array 함수를 이용하여 배열을 생성하고 x에 저장한다.
# 그 후에 x를 출력한다.
```

```python
from scipy import sparse
# scipy 모듈에서 sparse만 가져온다.
# scipy는 과학계산용 함수

eye =np.eye(4)
print("NumPy 배열 : \n", eye)4
# np 모듈의 eye 함수(대각 원소만 1이고 나머지 원소는 0인 배열)을 사용한다.
# 이런 함수를 sparse Matrix(희소 행렬 - 대부분의 원소가 0)이라 한다.
# eye 안의 숫자는 배열의 크기를 결정한다.
# 그 후에 eye를 출력한다.
```

```python
sparse_matrix = sparse.csr_matrix(eye)
print("\nScipy의 CSR 행렬 : \n",sparse_matrix)
# 위에서 만든 배열을 CSR 행렬로 만듦
# CSR 행렬이란? 행의 인덱스를 압축하여 저장
```

```python
sparse_matrix.toarray()
# 위에서 만든 CSR 행렬을 다시 배열로 바꿈
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

# 훈련하고 테스트로 분할을 하기 위해서 sklearn.model_selection 모듈의train_test_split를 가져옴
# data는 X, target은 y로 데이터들을 섞어서 train, test(75:25) 나눠줌
# random_state는 무작위에 seed를 주어 실험때마다 다른 값이 되는 것을 방지
```

```python
print("X_train 크기 : {}".format(X_train.shape))
print("y_train 크기 : {}".format(y_train.shape))
# 75:25로 나눴으므로 150의 75%인 112개가 됨
# X는 배열이므로 4열까지 있는 것까지 표현을 해줌
# 행렬은 대문자, 레이블/타겟은 소문자
```

```python
print("X_test 크기 : {}".format(X_test.shape))
print("y_test 크기 : {}".format(y_test.shape))
# 150의 25%인 38개가 출력됨
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# X_train 데이터를 이용해서 데이터 프레임 생성
# 열은 iris_dataset.feature_names을 사용

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', 
                          hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
# scatter는 산점도(점으로 표현)로 만들어줌
# s는 마커의 크기, cmap은 color map, alpha는 얼마나 투명할 것인가
# 대각선은 각 특성의 히스토그램
# 산점도는 아래 부분은 명확하게 되는데 윗 부분은 섞여있는 부분이 있음
# 대체적으로는 구분이 잘 됨
```

```python
from sklearn.neighbors import KNeighborsClassifier
# knn 분류기(KNeighborsClassifier)를 불러옴

knn = KNeighborsClassifier(n_neighbors=1)
# 이웃의 수를 1개로 설정(보통 홀 수로 설정, 짝수는 수가 같을 수 있음)

knn.fit(X_train, y_train)
# 훈련데이터와 레이블로 모델 훈련
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)
# 새로운 데이터를 생성해서 shape를 출력
```

```python
prediction = knn.predict(X_new)
print("예측:", prediction)
print("예측한 타깃의 이름:",
     iris_dataset['target_names'][prediction])
# 생성한 X_new를 knn의 prediction함수로 해서 prediction에 저장
# 3개의 클래스(0, 1, 2) 중에서 0으로 예측이 됨
# 0에 해당하는 타깃의 이름을 출력
```

```python
y_pred = knn.predict(X_test)
print("테스트 세트에 대한 예측값:\n", y_pred)
# X_test로 나눠놓은 데이터를 knn의 prediction 함수로 예측해서 y_pred에 저장
# 출력을 해보면 38개가 0, 1, 2로 예측된 것이 보임
```

```python
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test)))
# y_pred와 y_test와 비교해서 두 개가 같으면 같은애들 끼리 더해서 개수로 나눠서 정확도를 구함
```

```python
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))
# 다른 방법으로 X_test와 y_test를 score함수에 집어 넣어서 계산할 수도 있음
# 엄청 정확함
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1, 11)
# n_neigbor를 1부터 10까지 적용

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    # 이웃의 수를 1~10까지 적용
    clf.fit(X_train, y_train)
    # train 데이터로 훈련
    training_accuracy.append(clf.score(X_train, y_train))
    # train 정확도를 저장
    test_accuracy.append(clf.score(X_test, y_test))
    # 일반화 정확도를 저장
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
# cancer를 knn으로 출력
```
